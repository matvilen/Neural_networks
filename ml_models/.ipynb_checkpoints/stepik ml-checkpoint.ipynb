### Матан

**Поиск производной**

```python
# Определяем символьную переменную
x = sp.symbols('x')

# Определяем функцию
f = 3*x**2 - 12*x + 4

# Первая производная
f_prime = sp.diff(f, x)
print(f"f'(x) = {f_prime}")

# Вторая производная
f_double_prime = sp.diff(f, x, 2)
print(f"f''(x) = {f_double_prime}")

# Вычисление производной в точке
value_at_point = f_prime.subs(x, 0)
print(f"f'(2) = {value_at_point}")
```

**Приравниваем производную к нулю**

```python
equation = sp.Eq(f_prime, 0)
print(f"Уравнение: {equation}")

# Решаем уравнение
solutions = sp.solve(equation, x)
print(f"Критические точки: {solutions}")
```

**Пример 2**

```python
x = sp.symbols('x')
f = x*(x**2 - 9)
f_prime = sp.diff(f, x)
f_prime
```

```python
equation2 = sp.Eq(f_prime, 0)
solutions2 = sp.solve(equation2)
print(equation2, solutions2, sep='\n')
```

**Реализуем график**

```python
y = x*(x**2 - 9)  # y = x³ - 9x

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=0, color='k', alpha=0.3)
plt.axvline(x=0, color='k', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x(x² - 9)')
plt.show()
```
